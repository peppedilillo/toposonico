# Training data checks

```python
from pathlib import Path
import pandas as pd

data_path = Path("..") / "data/metadata/raw/training_pop30_genres.parquet"
df = pd.read_parquet(data_path)
print(f"Loaded {len(df):,} tracks from {data_path}")
```

## Leading artist

Since we are collapsing multiple tracks artists into one, we checks the logic is behaving as expected: the leading artist is the only artist for all tracks in an album.

```python
def consistent_first_artist(album_title: str, album_artist: str) -> bool:
    album_df = df[(df["album_name"] == album_title) & (df["artist_name"] == album_artist)]
    return set(df[df["id_upc"].isin(album_df["id_upc"])]["artist_name"]) == {album_artist}


ALBUMS_ARTISTS = [
    ("A Girl Like Me", "Rihanna"),
    ("Watch The Throne", "JAY-Z"),
    ("EVERYTHING IS LOVE", "The Carters"),
    ("An Evening With Silk Sonic", "Bruno Mars"),
    ("ASTROWORLD", "Travis Scott"),
    ("The Life Of Pablo", "Kanye West"),
    ("SOS", "SZA"),
    ("Random Access Memories", "Daft Punk"),
    ("Views", "Drake"),
    ("Humanz (Deluxe)", "Gorillaz"),
]

all([consistent_first_artist(t, a) for t, a in ALBUMS_ARTISTS])
```
