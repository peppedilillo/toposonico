## Sample Data

Fetch random samples from key tables.

```python
def sample_table(
    conn: sqlite3.Connection,
    table: str,
    n: int = 1000,
) -> pd.DataFrame:
    """
    Get a random sample from a table using ROWID-based sampling.

    This is fast even on huge tables because it:
    1. Gets max ROWID (uses index, instant)
    2. Generates random ROWIDs in Python
    3. Fetches only those specific rows
    """
    # Get max rowid
    cursor = conn.execute(f"SELECT MAX(rowid) FROM {table}")
    max_rowid = cursor.fetchone()[0]

    if max_rowid is None or max_rowid == 0:
        return pd.DataFrame()

    # Generate random rowids (sample more than needed to account for gaps)
    sample_size = min(n * 3, max_rowid)
    random_rowids = random.sample(range(1, max_rowid + 1), sample_size)

    # Fetch rows with those rowids
    placeholders = ",".join(["?"] * len(random_rowids))
    query = f"SELECT * FROM {table} WHERE rowid IN ({placeholders}) LIMIT {n}"
    return pd.read_sql_query(query, conn, params=random_rowids)
```

## Query Tracks with Audio Features

Since we have separate databases, we fetch track IDs first, then query audio features.

```python
# Get some tracks
tracks = query_to_df(
    spotify_conn,
    "SELECT id, name, popularity, duration_ms FROM tracks LIMIT 1000"
)

# Get audio features for those track IDs
track_ids = tuple(tracks["id"].tolist())
placeholders = ",".join(["?"] * len(track_ids))

audio_features = query_to_df(
    audio_conn,
    f"SELECT * FROM track_audio_features WHERE track_id IN ({placeholders})",
    track_ids
)

# Merge (audio uses track_id, tracks uses id)
tracks_with_features = tracks.merge(audio_features, left_on="id", right_on="track_id")
print(f"Shape: {tracks_with_features.shape}")
tracks_with_features.head(10)
```

```python
# Basic statistics of audio features
feature_cols = ["tempo", "energy", "danceability", "valence", "acousticness", "instrumentalness"]
available_cols = [c for c in feature_cols if c in tracks_with_features.columns]
tracks_with_features[available_cols].describe()
```

## Custom Queries

Use this cell to write your own queries.

```python
# Example: Find high-energy dance tracks
high_energy = query_to_df(
    audio_conn,
    """
    SELECT track_id, energy, danceability, tempo
    FROM track_audio_features
    WHERE energy > 0.8 AND danceability > 0.8
    LIMIT 100
    """
)

# Get track names
if not high_energy.empty:
    track_ids = tuple(high_energy["track_id"].tolist())
    placeholders = ",".join(["?"] * len(track_ids))
    track_names = query_to_df(
        spotify_conn,
        f"SELECT id, name FROM tracks WHERE id IN ({placeholders})",
        track_ids
    )
    result = high_energy.merge(track_names, left_on="track_id", right_on="id")
    result[["name", "energy", "danceability", "tempo"]].head(20)
```
